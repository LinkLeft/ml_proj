```python
import pandas as pd
import matplotlib.pyplot as plt

def process_stock_data(filepath):
    # 
    df = pd.read_excel(filepath)

    #  datetime 
    df['Names Date'] = pd.to_datetime(df['Names Date'])

    #  2020  9  1  2022  12  30 
    mask = (df['Names Date'] >= '2020-09-01') & (df['Names Date'] <= '2022-12-30')
    df = df.loc[mask]

    #  PERMNO 
    grouped = df.groupby('PERMNO')

    #  DataFrame 
    output_df = pd.DataFrame()

    # 
    for permno, group in grouped:
        # 1.  open price 
        plt.figure(figsize=(10, 6))
        plt.plot(group['Names Date'], group['Open Price'], marker='o', linestyle='-')
        plt.title(f'Open Price for PERMNO {permno} (2020-09-01 to 2022-12-30)')
        plt.xlabel('Date')
        plt.ylabel('Open Price')
        plt.grid(True)
        plt.show()

        # 2.  t+1 - t  open price 
        group['Open Price Change'] = group['Open Price'].diff()

        # 3. 
        print(f"\nDescriptive Statistics for PERMNO {permno}:")
        print(group['Open Price Change'].describe())

        #  DataFrame
        output_df = pd.concat([output_df, group])

    # 4.  Excel 
    output_df.to_excel("processed_stock_data.xlsx", index=False)

    print("\nData processing completed. Processed data saved to 'processed_stock_data.xlsx'.")


process_stock_data('DOW30_RAW_1.xlsx') 
```

```python
import pandas as pd

def classify_price_change(filepath):
    # 
    df = pd.read_excel(filepath)

    #  datetime 
    df['Names Date'] = pd.to_datetime(df['Names Date'])

    #  2020  9  1  2022  12  30 
    mask = (df['Names Date'] >= '2020-09-01') & (df['Names Date'] <= '2022-12-30')
    df = df.loc[mask]

    #  Open Price Change / Open Price 
    df['Price Change Ratio'] = df['Open Price Change'] / df['Open Price']

    # 
    def classify_ratio(ratio):
        if ratio < -0.01:
            return -1
        elif ratio > 0.01:
            return 1
        else:
            return 0

    df['Class'] = df['Price Change Ratio'].apply(classify_ratio)

    #  Excel 
    output_file = "classified_stock_data.xlsx"
    df.to_excel(output_file, index=False)

    print(f"Data processing completed. Classified data saved to '{output_file}'.")


classify_price_change('processed_stock_data.xlsx') 
```

```python
import pandas as pd

def filter_stock_data_by_date(filepath):
    # Excel
    df = pd.read_excel(filepath)
    
    # 'Names Date'datetime
    df['Names Date'] = pd.to_datetime(df['Names Date'])
    
    # 20209120221230
    start_date = '2020-09-01'
    end_date = '2022-12-30'
    mask = (df['Names Date'] >= start_date) & (df['Names Date'] <= end_date)
    filtered_df = df.loc[mask]
    
    # Excel
    output_filepath = 'train_stock_data.xlsx'
    filtered_df.to_excel(output_filepath, index=False)
    print(f" {output_filepath}")


filter_stock_data_by_date('DOW30allRAW.xlsx')
```
